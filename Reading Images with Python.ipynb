## Using Pillow to Read and Write Images

```python
from PIL import Image
```

```python
mandrill = Image.open("mandrill.png")
```

```python
mandrill
```

### Getting Image Info

```python
print("Image Format:", mandrill.format) # Image Format
print("Image Size(Width, Height):", mandrill.size) # Image Size
print("Image Mode:", mandrill.mode) # Image Mode
```

## Converting PIL Image to numpy array

```python
import numpy as np
mandrill_array = np.asarray(mandrill)
```

```python
print(type(mandrill_array))
print(mandrill_array)
```

## Using Matplotlib

```python
from matplotlib import image
import matplotlib.pyplot as plt
```

```python
mandrill = image.imread("mandrill.png")
plt.imshow(mandrill)
```

```python
print("Data type:", mandrill.dtype)
print("Shape:", mandrill.shape)
```
